# 03 第一个无未来函数回测

信号由当日收盘及过去数据计算，最早在下一交易日开盘执行。

```python
from investkb.cli import _demo_bars
from investkb.strategies import moving_average_signal
from investkb.backtest.engine import run_backtest
from investkb.backtest.models import FeeModel, MarketRules

bars = _demo_bars()  # 合成数据，不是真实行情
signal = moving_average_signal(bars['close'], fast=20, slow=60)
result = run_backtest(bars, signal, 100_000, FeeModel(), MarketRules())
result.trades.head()
```

检查第一笔交易日期：它必须晚于首次有效买入信号。随后再检查费用、不可成交、样本外和基准。
